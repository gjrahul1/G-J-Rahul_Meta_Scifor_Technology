```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv(r"/content/gdrive/MyDrive/Colab Notebooks/Datasets/titanic.csv")
df.head(10)
```

```python
df.drop(["PassengerId","Embarked","Cabin","SibSp","Parch","Ticket","Name"],axis="columns",inplace=True)
```

```python
df.head(7)
```

```python
target_variable = df.Survived
input_variable = df.drop('Survived',axis='columns')
```

```python
dummies = pd.get_dummies(input_variable.Sex,dtype=int)
dummies.head(5)
```

```python
input_variable = pd.concat([input_variable,dummies],axis='columns')
input_variable.head(5)
```

```python
#Droping off the duplicate columns
input_variable = input_variable.T.drop_duplicates().T
```

```python
input_variable = input_variable.drop("Sex",axis="columns")
```

```python
input_variable.head(10)
```

```python
input_variable.isnull().sum()
```

The Age has 177 null values

```python
input_variable.Age = input_variable.Age.fillna(input_variable.Age.mean())
input_variable.head(7)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input_variable,target_variable,test_size=0.3, random_state=42)  # Adjust test_size and random_state as needed
```

* The above method has object of pandas
* <b> x_train </b>, <b>y_train</b>, <b>x_test</b>, <b>y_test</b> are all pandas object
* If you are trying to replicate this notebook, then try adjusting the test_size if you face error in model.score

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
# Assuming y_train is your target variable
# Flatten y_train to 1-dimension if it's multi-dimensional
y_train = y_train.to_numpy().flatten() if y_train.ndim > 1 else y_train

# Convert y_train to numeric, handling errors
y_train = pd.to_numeric(y_train, errors='coerce')
y_train = y_train.astype(int)  # Convert to integers


# Now you can fit the model
model.fit(x_train, y_train[:x_train.shape[0]]) # slicing y_train to match x_train shape
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
score = model.score(x_test, y_test)
print(f"Score: ",score)
```
