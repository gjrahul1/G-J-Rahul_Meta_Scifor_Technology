```python
#Describe the data
print(df.describe())
```

```python
# Fill missing values in 'Declaration Date', 'Record Date', and 'Payment Date' with the next valid value
df['Declaration Date'].fillna(method='bfill', inplace=True)
df['Record Date'].fillna(method='bfill', inplace=True)
df['Payment Date'].fillna(method='bfill', inplace=True)

# Print the updated DataFrame to check if the missing values have been filled.
print(df.isnull().sum())
```

```python
df.drop_duplicates(inplace=True)
print(df.describe())
```

```python
print(df.head())
```

```python
# Extract the year from the 'Date' column
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
```

```python
#Group byy year and sum the dividends
yearly_dividends = df.groupby('Year')['Dividend'].sum()
```

```python
ax = yearly_dividends.plot.bar(
    figsize=(10, 7),
    color='skyblue',
    title='Distribution of Dividends Over the Years'
)
ax.set_xlabel('Year')
ax.set_ylabel('Total Dividends')
plt.xticks(rotation=45)
plt.show()
```

```python
yearly_dividends.plot.pie(autopct='%1.1f%%', startangle=90, cmap='viridis', figsize=(10, 7), title='Distribution of Dividends Over the Years')
```
