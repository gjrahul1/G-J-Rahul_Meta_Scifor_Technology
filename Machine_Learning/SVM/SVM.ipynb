```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
#Describe the dataset
dir(iris)
```

```python
iris.feature_names
```

```python
#Creating dataframe
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head(10)
```

```python
df['target'] = iris.target
df.head(10)
```

* We have added target thus making the dataframe complete for our analysis

```python
iris.target_names
```

## What are target_names ?
* Target names are those whom we consider as target in our analysis.
* In iris dataset, we have 3 target variables
  1. <b>Setosa</b>
  2. <b>Versicolor</b>
  3. <b>Virginica</b>



```python
#Versicolor
df[df.target==1].head(5)
```

```python
#Virginica
df[df.target==2].head(5)
```

```python
#Addition new column
df['flower_name'] = df.target.apply(lambda x:iris.target_names[x])
df.head(5)
```

* We are creating a new column out of already existing one ['target']
* Therefore in pandas we can use apply()
* Later, we use lambda function to add flower name to the respective target

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#Creating objects for each target
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]

#Checking the data type
print(f"df0: {type(df0)}")
print(f"df1: {type(df1)}")
print(f"df2: {type(df2)}")
```

## Data Visualization


```python
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color="green",marker='*')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color="blue",marker="+")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm")
plt.title("Visualizing Iris Dataset - Speal")
```

```python
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color="green",marker='*')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color="blue",marker="+")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm")
plt.title("Visualizing Iris Dataset - Petal")
```

### Data Training

```python
from sklearn.model_selection import train_test_split
X = df.drop(['target','flower_name'], axis=1)
Y = df.target
```

```python
X.head(5)
```

```python
Y.head(5)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC(C=2,gamma=1)
```

```python
model.fit(x_train,y_train)
```

```python
#Accuracy of the model
model.score(x_test,y_test)
```
