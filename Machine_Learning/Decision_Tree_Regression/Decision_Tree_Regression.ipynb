```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
load_data = pd.read_csv(r'C:\Users\gjrah\Documents\Data Science\Meta_Scifor\Data Sets\advertising.csv')
```

```python
load_data
```

We can observe that the dataset contains numerical categorical data.

<b> It contains the following categories: </b>
* TV
* Radio
* Newspaper
* Sales

```python
adv_data = load_data
```

```python
adv_data
```

```python
X = adv_data.drop(columns=['Sales'])
Y = adv_data['Sales']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeRegressor
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.6,random_state=20)
```

```python
X_train
```

```python
Y_train
```

<b> Convert the one-dimensional data to two-dimensional data </b>

```python
Y_train = np.array(Y_train).reshape(-1,1)
Y_train
```

```python
#Make sure the size of train and test variables is same.
#This enables it to Linear Regression

print(X_train.shape)
print(Y_train.shape)
```

```python
clf = DecisionTreeRegressor(random_state=42)
```

```python
clf.fit(X_train,Y_train)
```

<b> Prediction using Decision Tree Regressor </b>


```python
y_pred_train = clf.predict(X_train)
print(f"Predicted Sales: {y_pred_train}")
```

<h3> Plotting the graph</h3>

```python
# Assuming you have other features in adv_data
mean_values = adv_data.mean().drop('Sales')  # Drop the target variable
x_grid = np.linspace(adv_data['TV'].min(), adv_data['TV'].max(), 80).reshape(-1, 1)
x_grid_full = np.tile(mean_values.values, (x_grid.shape[0], 1))  # Repeat mean values for other features
x_grid_full[:, 0] = x_grid.flatten()  # Replace the first column with the grid for 'TV'
```

```python
y_pred_grid = clf.predict(x_grid_full)
```

```python
plt.scatter(adv_data['TV'], adv_data['Sales'],color='red')
plt.plot(x_grid,y_pred_grid,color="blue")
plt.title("TV vs Sales")
plt.xlabel("TV")
plt.ylabel("Sales")
plt.show()
```
