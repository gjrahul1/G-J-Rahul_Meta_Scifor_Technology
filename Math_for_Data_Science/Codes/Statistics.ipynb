<b> Quantile Range </b>

* 25% of the data
* 50% of the data
* 75% of the data
* <b>25% - 75% of the data is known as Inter-Quantile Range </b>

In the above data,

* 25% of the data lies at 5.5
* 50% of the data lies at 7.0
* 75% of the data lies at 7.75


```python
#Describing the quantile range
print(f"25%: {df.income.quantile(0.25)}\n 50%: {df.income.quantile(0.50)}\n 75%: {df.income.quantile(0.75)}")
```

```python
#Couple of ways to interpret quantiles
print(f"25% Lower:",df.income.quantile(0.25,interpolation="lower"))
print(f"25% Higer:",df.income.quantile(0.25,interpolation="higher"))
```

5000 (Lower) ──── 5500 (Linear) ──── 6000 (Higher)

```python
# Removing the outlier
# Before we reomve the outlier, we will understand where the outlier is present
# By viewing the data we can suggest, 6th row is an outlier
```

```python
identify_the_value = df.income.quantile(0.90)
identify_the_value
```

```python
#We can see at quantile range 90%, it has higher income, an noise!
df = df[df.income<identify_the_value]
df
```

* We have successfully pre-processed the data by removing the outlier, which now could be used for analytical purpose
